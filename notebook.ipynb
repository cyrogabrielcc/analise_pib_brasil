```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_excel('Dados_Pib.xlsx')
db
```

```python
# Agrupando a média por estado e ano
pib = db.groupby(by=['Territorialidades', 'Ano']).mean()
pib
```

```python
# Criando um relatório

# Cor de fundo
background = '#F5F5F5'

# CRIANDO O SISTEMA DE GRIDS
grid = sns.FacetGrid(db, col='Territorialidades',hue = 'Territorialidades', col_wrap=4)
grid = grid.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar sombra
grid = grid.map(plt.fill_between, 
                'Ano', 
                'PIB per capita', 
                alpha = 0.2).set_titles('{col_name} Territorialidades')

# Removendo as territorialidades
grid = grid.set_titles('{col_name}')

# Evolução da Renda per capta por estado
grid = grid.fig.suptitle('Evolução da Renda Per Capta por Estado', fontsize = 18)

# Ajustando os tamanhos
plt.subplots_adjust( top=0.92 )
```
